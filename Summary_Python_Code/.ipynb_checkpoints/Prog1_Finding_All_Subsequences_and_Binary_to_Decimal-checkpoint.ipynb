## Finding All Subsequences

```python
import itertools
list(itertools.combinations('01110',2))
```

```python
binary='001010110'
lst=[]
for i in range(1,len(binary)+1):
    lst.extend(list(itertools.combinations(binary,i)))
```

```python
lst
```

```python
tmp=[]
for i in range(len(lst)):
    tmp.append(''.join(lst[i]))
```

```python
tmp
```

```python
result=[]
for i in tmp:
    result.append(int(i, base=2))
```

```python
result=list(set(result))
result
```
